connect postgresql and import db

```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import dotenv_values

config = dotenv_values()

url = f"postgresql://{config['POSTGRES_USER']}:{config['POSTGRES_PASS']}@{config['POSTGRES_HOST']}:{config['POSTGRES_PORT']}/{config['POSTGRES_DB']}"
engine = create_engine(url)

query = """
SELECT *
FROM aero_insights.prep_final
ORDER BY date, airport_code
"""

df = pd.read_sql(query, engine)
df.shape
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# stats for flights

df.groupby("airport_code")[[
    "total_flights",
    "total_cancelled_num",
    "origin_cancelled_rate",
    "dest_cancelled_rate",
    "avg_dep_delay",
    "avg_arr_delay"
]].agg(["mean", "std", "min", "max", "sum"]).T
```

```python
df.groupby("airport_code")[[
    "avg_temp_c",
    "min_temp_c",
    "max_temp_c",
    "precipitation_mm",
    "max_snow_mm",
    "avg_wind_speed",
    "avg_peakgust"
]].agg(["mean", "std", "min", "max", "sum"]).T
```
